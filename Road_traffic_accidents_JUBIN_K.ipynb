```python
# 3. Feature engineering
cat
```

```python
df.dropna(subset=['Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality','Fitness_of_casuality'],inplace=True)
```

```python
na=df.columns[df.isna().any()].tolist()
na
```

```python
for i in na:
  df[i].fillna(df[i].mode()[0],inplace=True)
df.isna().sum()
```

```python
#frequency encoding
for i in cat:
  df_frequency=df[i].value_counts().to_dict()
  df[i]=df[i].replace(df_frequency)
  print(df[i])
```

```python
df
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
y
```

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
x,y=oversample.fit_resample(x,y)
```

```python
from sklearn.feature_selection import chi2
chi_sc=chi2(x,y)
chi_sc
```

```python
chi_scores=pd.DataFrame({'Features':x.columns,'Fscore':chi_sc[0],'Pvalues':chi_sc[1]})
chi_scores
chi_scores.sort_values(by='Fscore',ascending=False)
#small p values and higher fscore needed
```

```python
to_drop=chi_scores.loc[chi_scores['Fscore']<1000]['Features']
to_drop
df.drop(to_drop,axis=1,inplace=True)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
x_train
x_test
y_train
y_test
```

```python
#4. Model creation
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#when u have an imbalance dataset,that is one output outweighs other outputs
#we consider recall precision and f beta(f1 score)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
matrix=confusion_matrix(y_pred,y_test)
report=classification_report(y_pred,y_test)
score
```
